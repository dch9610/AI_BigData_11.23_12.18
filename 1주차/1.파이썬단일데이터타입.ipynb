# 파이썬-단일데이터타입

- 수치형(정수, 부동소수), 블린형, 문자열
- 변수
  - 값을 담는 그릇
  - 단, 파이썬에서는 데이터의 실체를 가르키는 주소를 가진 그릇이다 => **파이썬은 모든것이 객체이다**, 객체를 표현하는 변수는 주소만 가진다


```
a = 1
# a는 1을 가지고 있다(x), a는 1을 가르키고(참조하고) 있다(주소값을 가지고 있다)
a
```

```
# 객체를 확인하는 방법
# 도트 연산자 활용
# a.를 입력하고 잠시 기다리면 인텔리전스가 작동하여 후보들이 자동을 나타난다(객체다)
# 만약 아나콘다 주피터를 사용한다면 .을 입력후 TAB키를 누르면 나타난다
```

```
# a는 1을 참조하고 있다가, 
a = 1.1
# a는 1.1을 참조하게 변경되었다
a
```

```
# 왜 가능한가?
# 1. 파이썬은 타입 추론형 언어 => 값이 변수에 세팅될때 타입이 결정된다
# <-> 자바/c++/C# int a => 정수만와야한다
a = 1 # 값이 들어갈대 결정 -> 파이썬, js, 코틀린, 스위프트,...
# 2. 파이썬은 모든것이 객체이다. 1에서 1.1로 바뀐것은 단지 참조값만 변경됬다(주소변경만)
# a는 정수 ok, 부동소수ok, 문자열 ok, 함수ok, 객체 ok... 다 된다!!
# 따라서, 사전에 정의된 클레스, 내장함수, 외장함수등등 이름을 변수로 사용하면 않된다
```
