```
# 7. 인덱싱(중요!) -> 데이터중 원하는 부분을 추출
# 차원축소를 전제로 가진다
# ex) 2차배열의 인덱싱 1차배열이 결과
# ex) 백터를 인덱싱하면 스칼라가 나온다
# ex) DataFrame을 인덱싱하면 Series가 나온다
# 문법:변수명[인덱스(0,1,2,..:정방향,역방향(..-3,-2,-1))]
a = '0123456789'
a
```

```
# a 문자열에서 3을 추출하시오->인덱싱
a[3], a[-7]
# 커서의 오른쪽값을 취한다
# 인덱스 방향은 가까운쪽으로 카운트한다
```

```
# 8. 슬라이싱(중요!) -> 데이터중 원하는 부분을 자른다
# 자른다 => 차원유지
# 문법:변수[시작인덱스:종료인덱스:step]
# step:간격(생략하면 1)
a
```

```
# 시작인덱스:종료인덱스가 생략되면, 처음부터~ 끝까지
# 원본 카피
# x:y => x<= 값 <y
a[:], a[::2], a[1:-1], a[:2], a[-2:]
```

```
# 11. 포멧팅
# 문자열과 문자열의 결합, 문자열안에 특정 데이터를 삽입하여 생성
# 11-1. 문자열 % (파라미터)
# 11-2. 문자열.foramt(파라미터)
# 11-3. f'문자{파라미터}열' <- 가장 최근에 추가되었다 3.6부터 사용가능
# 데이터를 원하는 형태로 가공 처리 위해서 사용 -> 정제의 역활
# 향후 문자열 전처리 => 정규식 처리(단독 주제)
```

```
a = 1
b = 2
```

```
# 타입이 맞지 않아서 에러
a + " + " + b + '=' + (a+b)
```

```
str(a) + " + " + str(b) + ' = ' + str(a+b)
```

```
# 포멧팅 => 결과를 먼저 쓰고 나머지는 채운다
"%d + %d = %d" % (a, b, a/b)
```

```
# 11-1
# 위의 표현은 타입을 제한적으로 사용해서 타입이 틀티면 잠재적인 오류
# 예상된다 -> 범용적으로 표현하여 에방
"%s + %s = %s" % (a, b, a/b)
# %s는 입력값의 타입이 뭔지 몰라도 다 처리해준다
# 단, 순서를 바꿀수 없다, 순서대로 자리에 들어간다
```

```
# 11-2
# 위의 한계점등을 개선
"{} + {} = {}".format(a, b, a/b)
```

```
"{0} + {1} = {2}".format(a, b, a/b)
```

```
# 순서를 바꿔도 ok
"{1} + {0} = {2}".format(a, b, a/b)
```

```
# 파라미터를 사용하지 않아도 ok
"{1} + {1} = {0}".format(a, b, a/b)
```

```
# 의미를 명확하게 부여할수 있다
"{height} + {weight} = {result}".format(height=a, weight=b, result=a/b)
```

```
# 11-3
# f-string 포멧팅, 가장 간결
f'{a}+{b}={a+b}'
```

```
# 12. 공백제거
tmp = '''
				
					
					
					
					유럽연합 EUR
				
'''
# 문자열의 앞뒤 공백 제거 -> 문자열 함수 사용
f'[{tmp.strip()}]', f'[{tmp.lstrip()}]', f'[{tmp.rstrip()}]'
```

```
# 문자열의 내부 공백 제거, 복잡한 패턴의 데이터를 필터링, 추출
# 정규식으로 해결!!
# 간단한것은 문자열 함수들로 처리하고, 복잡한 것은 정규식으로 처리
```

```
# 14. 문자열 포멧팅 응용에서 값의 자리수(칸수) 지정
'[%s]' % 'abcd'
```

```
# 10칸을 무조건 차지한다
'[%10s]' % 'abcd', '[%-10s]' % 'abcd'
```

```
'[%10f]' % 3.145678, '[%10s]' % 3.145678
```

```
'[%10.2f]' % 3.145678
```

```
'[{:10}]'.format('abcd')
# '[{0:-10}]'.format('abcd') : 에러 발생
```

```
# 앞쪽, 뒤쪽으로 정렬
'[{0:<10}]'.format('abcd'), '[{0:>10}]'.format('abcd')
```

```
# 중심
'[{0:^10}]'.format('abcd'), '[{0:*^10}]'.format('abcd')
```

```
a = 'abcd'
f'[{a:10}]', f'[{a:>10}]', f'[{a:^10}]'
```
