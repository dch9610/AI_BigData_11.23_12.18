## 기본 if문

```
# 포스에서 계산, 케이스는 현금결제모드 3개
# (과하게,딱맞게,부족하게 입금) => 케이스별로 액션이 다르다
# 조건에 맞춰서 분기해야 한다
# : => 코드블럭이 시작된다 => 타언어 {  statements,... }
if 조건: # 돈을 더 많이 주었을 경우
  # 파이썬의 들여쓰기는 탭으로 밀어주면된다
  # pass는 포지션을 차지하고는데, 아무일도 하지 않는다
  # 조건문, 반복문, 함수, 클레스등 틀을 잡을때 사용
  pass
elif 조건:
  pass
else:
  pass
```

```
# 상수처럼 취급하겟다. 환경변수
COFFEE_ICE_PRICE = 1500
```

```
# 비교연산자,  >, <, >=, <=, ==, !=
# 비트연산자, not, and, or
#coffee_price = 500
# 퍼포먼스를 고려하여 가장 많은 케이스를 먼저 표시
if coffee_price > COFFEE_ICE_PRICE:    # 돈을 더 많이 주었을 경우
  print(f'거스름돈이 {coffee_price-COFFEE_ICE_PRICE}원 입니다.')
  pass
elif coffee_price == COFFEE_ICE_PRICE: # 같은 가격을 입력받았다면
  print('안녕히 가세요')
  pass
else:# 부족하게 냈다
  print(f'부족한 금액이 {COFFEE_ICE_PRICE-coffee_price}원 입니다.')
  pass
```

```
# 비트 연산중 부정을 잡는 조건문 => not
# 않되는것, False 상황을 잡는 케이스
isRunning = False
# 코드블럭의 수행문이 1개이면 옆으로 나열할수 있다
# 한줄에 수행문에 ;이라는 구분자를 쓰면 여러개를 나열할수 있다
if isRunning:print('가동중');print('가동중')
else:print('중단중')
```

# for

- for ~ index 방식 미지원, for ~ each(in) 방식 지원
- 파이썬은 for ~ each만 지원한다

```
# 연속데이터 타입의 변수명은 복수로 묘사한다
items = [1,2,3,4,5]
# 한개씩 꺼내서 출력하시오
# 아이템스 안에서 아이템을 한개씩 순서대로 뽑아낸다
for item in items:
  print( item )
```

```
items = [(1,2),(3,4),(5,6)]
# 한개씩 뽑아서 출력하시오
for item in items:
  print( item )
# 값들을 직접 변수로 받아서 데이터를 획득하고 싶다
# 앞에값은 필요없고, 뒤에 값만 필요하다면
for _,b in items:
  print( b )
```

```
# 환율 데이터
exchanges = [
   {
      'code':'USD',
      'buy':1118.11
   },
   {
      'code':'EUR',
      'buy':1318
   }
]
exchanges
```

```
# 데이터를 돌면서, 
# ex) "USD를 구매할때, 한화로 1318원이 필요합니다." 이렇게 출력
for exchange in exchanges:
  # 딕셔너리에서 키를 넣어서 값을 추출 => 인덱싱
  # 인덱싱 => 변수[ 키 ]
  # 문자열이나 코드가 길어져서 내려친다면 \ 표현하여 이어짐을 표시
  print( f"{ exchange[ 'code' ] }를 구매할때, \
  한화로 {exchange[ 'buy' ]}원이 필요합니다." )
```

```
# range()를 활용하여 비트연산 ( 0 or 1)
# "0, 0, bit연산 -> 0 and 0, 0 or 0, 0 xor 0" <- 출력값
# "0, 1, bit연산 -> and, or, xor" <- 출력값
# "1, 0, bit연산 -> and, or, xor" <- 출력값
# "1, 1, bit연산 -> and, or, xor" <- 출력값
# 2중 for문 케이스 => 모든 케이스를 만들때
for n in range(2):   # 0,1
  for m in range(2): # 0,1
    print( f'n={n}, m={m} bit연산 -> {n and m},{n or m},{n ^ m}' )
```

```
# lv1 구구단 3단~5단까지 구현하시오
# 출력내용 "3 x 1 = 3, ...5 x 9 = 45"
# 앞자리 3~5, 뒷자리 1~9
for n in range(3, 5+1):
  for m in range(1, 9+1):
    print(f'{n} x {m} = {n*m:>2}')
```

```
# lv2 3단, 5단만 출력하시오
for n in range(3, 5+1, 2):
  for m in range(1, 9+1):
    print(f'{n} x {m} = {n*m:>2}')
```

```
# 결과는 모두 리스트에 모두 담는다
# 1. 담을 그릇을 준비=>리스트
results = list() #[]
for n in range(3, 5+1, 2):
  for m in range(1, 9+1):
    # 2. 담는다-> 리스트에 맴버로 추가한다
    results.append(f'{n} x {m} = {n*m:>2}')
# 3. 확인
len( results )
```

# while

- 0 ~ 무한대의 수행횟수를 가진 반복문
- 타언어는 1 ~ 무한대도 잇다. (do~ while)
- 언제 끝날지 모르겠다

```
a = [1,2,3,4,5]
a,len(a)
```

```
'''
  while 조건:
    여러줄의 수행문(statements)
'''
# a의 맴버가 없어질때까지 반복해서 => a의 맴버가 존재하면,맴버수가 0보다 크면
# 반복할때 각 맴버를 추출하여 출력한다
while len(a)>0:
  # 맴버를 뒤에서 혹은 앞에서 추출하여 출력하는 전제가 없으므로, 
  # pop() 사용해도 무방
  print( a.pop() )
```

```
# 파이썬이라면 좀더 줄일수 있지 않을까? => 조건문의 결론:True or False
# False로 판단되는 상황 => 0, 0.0, '', [], {}, (), None, NaN, 기타조건상황
# a는 정체가 => []
# 맴버가 존재할때까지라는 조건문이 나왔으므로
a = [1,2,3,4,5]
while a:
  print( a.pop() )
```

```
a = [1,2,3,4,5]
while a:
  print('위에서 조건 체크하고, 조건을 만족하면, 코드블럭이 수행된다')
  print(f'총맴버수 {len(a)}')
  print( a.pop() )
  print(f'총맴버수 {len(a)}')
  print('-'*30)
```

```
# 반복문이 중간에 "중단없이" 모두 계획되로 반복했는가? 체킹 -> else
# else는 for, while, try등에서 정상적으로 수행되었는지를 체킹하는 표현
a = [1,2,3,4,5]
while a:
  print( a.pop() )
else:
  print( '정상적으로 중간에 중단없이 반복문이 수행되었다는 것을 보증한다' )
```

- break
   - 통상 조건문과 함께, 혹은 단독으로 break를 사용하는 지점에서 반복문을 중단시킨다
   - break는 자기를 둘러싸고 있는 가장 가까운 반복문을 빠져나간다

```
a = [1,2,3,4,5]
while a:
  print( a.pop() )
  # 남은 맴버수가 2개가 되면 그만해라-> 중단
  if len(a) == 2:
    break
else:
  # 반복문이 중단되었기 때문에 여기로 진입하지 않는다
  # else 구문은 필요시 사용한다.
  print( '정상적으로 중간에 중단없이 반복문이 수행되었다는 것을 보증한다' )
```

- continue
  - 통상 조건문과 함께, 혹은 단독으로 continue를 사용하는 지점에서 반복문의 조건식으로 바로 점프한다
  - continue를 둘러싸고 있는 가장 가까운 반복문으로 점프
  - 점프후에는 바로 조건문 수행
  - while, for만 사용 가능

```
a = [1,2,3,4,5]
while a:
  tmp = a.pop() #변수 = a.pop()  
  if tmp % 2 == 0: # 짝수면
    continue
  print( tmp )  #print( 변수 )
  if len(a) == 1:
    break
else:  
  # 중간에 중단된적이 없으므로 출력된다(continue는 관계 없음)
  print( '정상적으로 중간에 중단없이 반복문이 수행되었다는 것을 보증한다' )
```
